```python
#importing python libraries, and identified the 5 features from the dataset
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as pl
```

```python
RuterData = pd.read_csv('ruters.csv',sep='\t' , delimiter = ';',  header=None, encoding ='latin1', names=['Tur_ID', 'Kommune', 'Linjetype', 'Faktisk_Ankomst', 'Planlagt_Ankomst', 'Kapasitet', 'Ombord'])
```

```python
#printing the rows
RuterData.head()
```

```python
#Storing data into different lists

x=RuterData["Ombord"]
y=RuterData["Kapasitet"]

#scattering the vehicle's capacity vs how passengers were on board
pl.scatter(x,y)
pl.xlabel('Plass_Kapasitet->')
pl.ylabel('Kapasitet->')
pl.title('Data')
pl.show()
```

```python
#Bar chart of how many passengers there were in the vehicles, and it is also showing how many times Ruter collects false data from the sensors
#for example: some data shows that there were -2 passengers, something that is not possible in real life
RuterData['Ombord'].value_counts().sort_index().plot.bar()
```

```python
#Another bar chart that is horizontal, this chart shows the capacity of the vehicles
RuterData['Kapasitet'].value_counts().sort_index().plot.barh()
```

```python
#Shows the counted value difference between the local, express, school and night vehicles
RuterData['Linjetype'].value_counts()
```

```python
sns.countplot(RuterData['Linjetype'])
```

```python
#Shows how many vehciles the different municipalities have access to 
sns.countplot(RuterData["Kommune"])
```
